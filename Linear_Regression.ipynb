```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
df=pd.read_csv('weight-height.csv')
```

```python
df.drop('Gender',axis=1,inplace=True)
```

```python
df.head()
```

```python
plt.scatter(df['Height'],df['Weight'])
plt.xlabel('Height')
plt.ylabel('Weight(in cm)')
```

```python
x=df.iloc[:,0:1]
y=df.iloc[:,-1]
```

```python
x
```

```python
from sklearn.model_selection import train_test_split
xtrain, xtest, ytrain, ytest = train_test_split(x, y, test_size=0.2,random_state=2)
```

```python
xtrain
```

```python
ytrain
```

```python
xtest
```

```python
ytest
```

```python
from sklearn.linear_model import LinearRegression
```

```python
lr=LinearRegression()
```

```python
lr.fit(xtrain,ytrain)
```

```python
xtest
```

```python
ytest
```

```python
lr.predict(xtest.iloc[0].values.reshape(1,1))
```

```python
plt.scatter(df['Height'],df['Weight'])
plt.plot(xtrain,lr.predict(xtrain),color='yellow');
plt.xlabel('Height')
plt.ylabel('Weight(in cm)')
```

```python
m=lr.coef_
```

```python
m
```

```python
b=lr.intercept_
b
```

```python
#y=mx+b
m*9.5+b
```
